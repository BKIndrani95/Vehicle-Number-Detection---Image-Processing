```python
# Importing Necessary Libraries
import os
import cv2
import pytesseract
import pandas as pd

# Providing Necessary files path
image_folder = r'C:\Users\HP\Downloads\Licplatesrecognition_train\license_plates_recognition_train'
csv_file = r'D:\Ind\Licplatesdetection_train.csv'
```

```python
# converting CSV file into DataFrame
df = pd.read_csv(csv_file)
extracted_text_list = [] # Creating list to add fetched text from image
```

```python
# using For loop to iterate
for index, row in df.iterrows():
    img_id = row['img_id']  
    img_id = img_id.replace(".jpg", "") #removing ".jpg" from img_id column
    image_path = os.path.join(image_folder, f"{img_id}.jpg")  # Append ".jpg" to the image ID
    print("Constructed image path:", image_path)
    if os.path.exists(image_path):
        print(f"Processing image: {image_path}")  # Print the path to the image being processed
        try:
            image = cv2.imread(image_path)
            if image is not None: # if it consists it should load
                print("Image loaded successfully.")
                gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                extracted_text = pytesseract.image_to_string(gray_image) # Extracting texts from Images
                extracted_text_list.append(extracted_text.strip())
            else:
                print(f"Failed to load image: {image_path}") # incase fails to load images it must show ""
                extracted_text_list.append("")  
        except Exception as e:
            print(f"Error processing image {image_path}: {e}") #Error Handling
            extracted_text_list.append("")  
    else:
        print(f"Image {img_id}.jpg not found.")
        extracted_text_list.append("")  
```

```python
output_df = pd.DataFrame({'img_id': df['img_id'], 'extracted_text': extracted_text_list}) # adding img_id and extracted_text data in DataFrame
print(output_df)
output_csv_file = r'D:\Ind\ExtractedText.csv' #csv file path
output_df.to_csv(output_csv_file, index=False)
print(f"Extracted text saved to {output_csv_file}") 
```
